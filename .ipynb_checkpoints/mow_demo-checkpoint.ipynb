```python
import maintsim
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# serial line no maintenance
line = maintsim.SimulationParameters(
    process_times = [4, 2, 5, 3, 2],
    buffer_sizes = [2]*5,
    #planned_failures = [(0, 50, 5)],#, (1, 30, 15), (2, 60, 10)],
    
    sim_time = 250, warmup_time=0)

scenario1 = maintsim.Scenario(line)

scenario1.simulate()
```

### Single planned failure on M0, critical downtime not exceeded

```python
# single downtime
# serial line CM of planned failures
line = maintsim.SimulationParameters(
    process_times = [4, 2, 5, 3, 2],
    sim_time = 100,

    buffer_sizes = [2]*5,
    
    degradation = [1.]*5,
    
    maint_policy = 'CM',
    repair_dist = 'uniform',
    repair_params = [5, 15],
    queue_discipline = 'fifo',
    maint_capacity = 1,
    
    planned_failures = [(0, 50, 20)],
    
    cm_cost = 100,
    
    warmup_time = 1000)

scenario1 = maintsim.Scenario(line)

scenario1.simulate()
```

```python
bottleneck_idle = pd.DataFrame(columns=[str(st) for st in range(50,56)], index=list(range(19,26)))
idle = []
for st in range(50,56):
    for dt in range(19,26):
        line = maintsim.SimulationParameters(
            process_times = [4, 2, 5, 3, 2],
            sim_time = 100,

            buffer_sizes = [2]*5,

            degradation = [1.]*5,

            maint_policy = 'CM',
            repair_dist = 'uniform',
            repair_params = [5, 15],
            queue_discipline = 'fifo',
            maint_capacity = 1,

            planned_failures = [(0, st, dt)],

            cm_cost = 100,

            warmup_time = 1000)

        scenario2 = maintsim.Scenario(line)
        
        bottleneck_idle.loc[dt, str(st)] = scenario2.simulate(summary=False, return_PMOW=True)
        
        #idle += [scenario2.simulate(summary=False, return_PMOW=True)]
print('Done.')
```

### Multiple planned failures on M0, M1

```python
# serial line CM of planned failures
line = maintsim.SimulationParameters(
    process_times = [4, 2, 5, 3, 2],
    sim_time = 100,

    buffer_sizes = [2]*5,
    
    degradation = [1.]*5,
    
    maint_policy = 'CM',
    repair_dist = 'uniform',
    repair_params = [5, 15],
    queue_discipline = 'fifo',
    maint_capacity = 1,
    
    planned_failures = [(0, 10, 30), (1, 75, 30)],
    
    cm_cost = 100,
    
    warmup_time = 1000)

scenario3 = maintsim.Scenario(line)

scenario3.simulate()
```

```python
machines = [0,1,2]
plt.figure(figsize=(12,8))
for i in machines:
    plt.plot(scenario3.production_data.loc[0:99,'Machine {} production'.format(i)],
        label='M{}'.format(i+1), lw=3)
plt.plot(scenario3.production_data.loc[0:99,'Ideal production'], c='grey', lw=3, alpha=0.5)
plt.plot(scenario3.production_data.loc[0:99,'M2 processing'], label='Bottleneck idle time', lw=3, c='red')
plt.xticks(np.arange(0,101,10), np.arange(0, 101, 10))
plt.xlabel('time', fontsize=14)
plt.ylabel('units produced', fontsize=14)
plt.legend(fontsize=14)
plt.title('Production under deterministic downtime', fontsize=14)
plt.grid()
#plt.savefig('C:/Users/Michael/Desktop/Temp/case2_production.eps', format='eps')
plt.show()
```
